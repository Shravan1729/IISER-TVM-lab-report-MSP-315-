```python
import numpy as np
import matplotlib.pyplot as plt
from math import atan, tan

def linear_regression(X, Y):
    """
    Perform linear regression on arrays of X and Y values.

    Parameters:
    X (array-like): Independent variable values.
    Y (array-like): Dependent variable values.

    Returns:
    slope (float): Slope of the regression line.
    intercept (float): Intercept of the regression line.
    """
    
    # Number of data points
    n = len(X)

    # Calculating the means of X and Y
    mean_X = np.mean(X)
    mean_Y = np.mean(Y)

    # Calculating the cross-deviation and deviation about X
    SS_xy = np.sum(Y*X) - n*mean_Y*mean_X
    SS_xx = np.sum(X*X) - n*mean_X*mean_X

    # Calculating the regression coefficients
    slope = SS_xy / SS_xx
    intercept = mean_Y - slope*mean_X

    cot_theta = 1/(tan(atan(slope)))

    # Predicting Y values
    Y_pred = slope * X + intercept



    return slope, intercept, cot_theta, Y_pred
```

#### *For Transverse Direction*

```python
X = np.array([i for i in range(2,22,2)])
Y_2 = np.array([1.37, 3.42, 5.24, 7.24, 8.94, 10.89, 13.40, 15.52, 17.47, 19.36])   # mass = 200 gms 
Y_1 = np.array([2.66, 5.57, 9.18, 13.25, 16.40, 19.98, 23.52, 27.04, 30.58, 31.14])   # mass = 150 gms 

slope_1, intercept_1, cot_theta_1, Y_pred_1= linear_regression(X, Y_1)
slope_2, intercept_2, cot_theta_2, Y_pred_2 = linear_regression(X, Y_2)

print("Slope and intercept for the m = 150 gms ")
print(f"Slope: {slope_1}")
print(f"Cot(theta): {cot_theta_1} ")

print("\nSlope and intercept for the m = 200 gms ")
print(f"Slope: {slope_2}")
print(f"Cot(theta): {cot_theta_2} ")

# Plotting the results
plt.figure(figsize=(16,7))

plt.subplot(121)
plt.grid()
plt.scatter(X, Y_1, color="blue", label="Original data for m = 150 gms")
plt.plot(X, Y_pred_1, color="red", label="Fitted line")
plt.xlabel(r"Order no. $(n)$")
plt.ylabel(r"Diameter squared $(D^2 _n)$")
plt.legend(loc='best')

plt.subplot(122)
plt.grid()
plt.scatter(X, Y_2, color="blue", label="Original data for m = 200 gms")
plt.plot(X, Y_pred_2, color="red", label="Fitted line")
plt.xlabel(r"Order no. $(n)$")
plt.ylabel(r"Diameter squared $(D^2 _n)$")
plt.legend(loc='best')

plt.savefig('cornu Transverse.png')
plt.show()
```
