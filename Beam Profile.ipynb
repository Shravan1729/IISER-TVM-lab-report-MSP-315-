```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.optimize import curve_fit
```

#### For Horizontal Direction

```python
# Experimental data
X_1 = np.array([1.99, 2.31, 2.37, 2.63, 2.69, 2.72, 2.78, 2.98, 3.13, 3.19, 3.43, 3.68])
Y_1 = np.array([2, 2.1, 2.2, 2.3, 2.4, 2.3, 2.2, 2.2, 2.1, 2.0, 1.9, 1.8])

# Define Gaussian function
def gaussian(x, amplitude, mean, stddev):
    return amplitude * np.exp(-((x - mean) ** 2) / (2 * stddev ** 2))

# Fit the Gaussian curve to the data
popt, _ = curve_fit(gaussian, X_1, Y_1, p0=[1, np.mean(X_1), np.std(X_1)])

# Generate X values for the fit curve
X_fit = np.linspace(min(X_1), max(X_1), 100)
Y_fit = gaussian(X_fit, *popt)

# Plot the data and the fitted Gaussian curve
plt.title("Beam Profile at Horizontal Direction")
plt.scatter(X_1, Y_1, label='Experimental Data', color='blue')
plt.plot(X_fit, Y_fit, label='Theoretical Fit', color='red', linestyle='--')
plt.xlabel('Distance (mm)')
plt.ylabel('Intensity (mA)')
plt.legend(loc='best')
plt.savefig("Beam Profile at Horizontal Direction.png")
plt.show()
```
