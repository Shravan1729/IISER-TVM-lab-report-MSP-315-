```python
import numpy as np
import matplotlib.pyplot as plt

'''Series LCR'''
# Frequency data
frequency_data = np.array([0.50, 0.60, 0.70, 0.80, 0.90, 1.00, 1.10, 1.20, 1.30, 1.40, 1.45, 1.50, 1.55, 1.60, 1.70, 1.80, 1.90, 2.00, 2.10, 2.20, 2.30])
# Voltage data
voltage_data = np.array([192.0, 210.0, 248.0, 304.0, 368.0, 464.0, 576.0, 704.0, 832.0, 960.0, 1008.0, 1024.0, 1016.0, 984.0, 904.0, 800.0, 704.0, 608.0, 560.0, 504.0, 464.0])

max_index = np.argmax(voltage_data)
res_freq = frequency_data[max_index]

'''Parallel LCR'''
import numpy as np

# Data from the image
frequency_data_2 = np.array([0.50, 0.60, 0.70, 0.80, 0.90, 1.00, 1.10, 1.20, 1.30, 1.40, 1.45, 1.50, 1.55, 1.60, 1.70, 1.80, 1.90, 2.00])
voltage_data_2 = np.array([760.0, 664.0, 548.0, 480.0, 392.0, 320.0, 248.0, 192.0, 132.0, 88.0, 60.0, 58.0, 63.0, 88.0, 123.0, 199.0, 260.0, 334.0])
min_index = np.argmin(voltage_data_2)
res_freq_2 = frequency_data_2[min_index]


#Plotting
plt.figure(figsize=(12,6))
plt.subplot(121)
plt.plot(frequency_data, voltage_data, '--')
plt.scatter(frequency_data, voltage_data, color='black', marker='+')
plt.axvline(res_freq, color = 'k', label='Resonant Frequency')
plt.grid()
plt.xlabel(r"Frequency $(kHz)$")
plt.ylabel(r"Voltage $(mV)$")
plt.legend(loc='best')
plt.title("Frequency vs. Voltage plot for Serial LCR")

plt.subplot(122)
plt.plot(frequency_data_2, voltage_data_2, '--')
plt.scatter(frequency_data_2, voltage_data_2, color='black', marker='+')
plt.axvline(res_freq,color = 'k', label='Resonant Frequency')
plt.grid()
plt.xlabel(r"Frequency $(kHz)$")
plt.ylabel(r"Voltage $(mV)$")
plt.legend(loc='best')
plt.title("Frequency vs. Voltage plot for Parallel LCR")


plt.savefig('Frequency_response_LCR.png')
plt.show()
```
