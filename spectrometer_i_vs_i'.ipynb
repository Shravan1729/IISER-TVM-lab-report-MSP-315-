```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

# Experimental data
i = np.array([35, 40, 45, 50, 55, 60, 65])
i_bar = np.array([68.95, 60.48, 54.03, 50.25, 45.00, 41.04, 37.75])

# Define the function for the 1/x fit
def fit_func(x, a, b):
    return a / x + b

# Fit the data using the 1/x function
params_curve, _ = curve_fit(fit_func, i, i_bar)
a, b = params_curve

# Define the straight line y = x
def line(x):
    return x

# Generate values for plotting
X = np.linspace(30, 70, 1000)
Y = fit_func(X, a, b)
Straight_line = line(X)

# Find the intersection by checking where the absolute difference is minimal
def find_intersection(X, Y, Straight_line):
    differences = np.abs(Y - Straight_line)
    min_diff_index = np.argmin(differences)
    return X[min_diff_index], Y[min_diff_index]

# Calculate the intersection point
intersection_x, intersection_y = find_intersection(X, Y, Straight_line)


# Plot the data
plt.plot(X, Y, label='Mean Curve',linestyle= '-.')
plt.plot(X, Straight_line, label='Straight line Passing through the Origin', linestyle='--')
plt.scatter(i, i_bar, color='k',marker='^', label='Data Points',s=60)
plt.scatter(intersection_x, intersection_y, color='magenta', zorder=5, marker='x', s = 100, label=f'Y intercept of the Curve is:  {intersection_y:.2f}$\degree$')

# Customize the plot
plt.xlabel(r'Angle of incidence ($i$)')
plt.ylabel(r"Angle of Emergence ($i'$)")
plt.title('Angle of Incidence vs Angle of Emergence Curve')
plt.legend(loc='best')
plt.grid(True)
plt.savefig('Angle of Incidence vs Angle of Emergence Curve.png')
plt.show()
```
